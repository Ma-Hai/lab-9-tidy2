```python
# https://www.kaggle.com/datasets/atharvasoundankar/global-energy-consumption-2000-2024 


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = pd.read_csv("dataset/energy.csv")

df = df.drop_duplicates()
df['Industrial Energy Use Amount']=df['Total Energy Consumption (TWh)']*df['Household Energy Use (%)']/100
df = df.sort_values(by='Country',ignore_index=True)
df.columns = map(str.lower, df.columns)
df.rename(columns=lambda col: col.replace(' ', '_'), inplace=True)
#No issues in dataset, just replaced ' ' with '_' and lowered title names.

df.head()

australia_df = df[df['country']=='Australia']
australia_df = australia_df.sort_values(by='year',ignore_index=True)
# australia_df = australia_df.drop_duplicates('year')
australia_df.head(20)


USA_df = df[df['country']=='USA']
USA_df = USA_df.sort_values(by='year',ignore_index=True)
# USA_df = USA_df.drop_duplicates('Year')
USA_df.head(20)

China_df = df[df['country']=='China']
China_df = China_df.sort_values(by='year',ignore_index=True)
# USA_df = USA_df.drop_duplicates('Year')
China_df.head(20)

plt.rcParams['font.size'] = 12

sns.set_theme(style='whitegrid', context='notebook', font_scale=2)
plt.figure(figsize=(40,10))
sns.barplot(data=USA_df, x='year', y='industrial_energy_use_amount', palette='Set2')
plt.title('USA')
plt.xlabel('Year')
plt.ylabel('Industrial Energy Use Amount')
sns.despine()
plt.show()

sns.set_theme(style='whitegrid', context='notebook', font_scale=2)
plt.figure(figsize=(40,10))
sns.barplot(data=China_df, x='year', y='industrial_energy_use_amount', palette='Set2')
plt.title('China')
plt.xlabel('Year')
plt.ylabel('Industrial Energy Use Amount')
sns.despine()
plt.show()


sns.set_theme(style='whitegrid', context='notebook',font_scale=2)

plt.figure(figsize=(40,10))
sns.barplot(data=australia_df, x='year', y='industrial_energy_use_amount', palette='Set2')
plt.title('Australia')
plt.xlabel('Year')
plt.ylabel('Industrial Energy Use Amount')
sns.despine()
plt.show()


sns.set_theme(style='whitegrid', context='notebook',font_scale=2)
plt.figure(figsize=(40,10))
sns.barplot(data=USA_df, x='year', y='carbon_emissions_(million_tons)', palette='Set2')
plt.title('USA')
plt.xlabel('Year')
plt.ylabel('Carbon Emissions (Million Tons)')
sns.despine()
plt.show()
```

USA and Australia Industrial Energy Usage both seem to peak at similar years such as 2002 and 2012 while China doesn't reflect as much of a similar trend. It is also interesting to note that despite Climate Change efforts, the US carbon emissions is fluctuating without significant indications of a decrease in emission.
